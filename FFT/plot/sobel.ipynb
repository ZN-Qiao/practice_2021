```python
import matplotlib.pylab as plt
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import filters
im = rgb2gray(imread('Places365_val_00004671.jpeg')) # RGB image to gray scale
plt.gray()
plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(im)
plt.title('original', size=20)
plt.subplot(222)
edges_y = filters.sobel_h(im) 
plt.imshow(edges_y)
plt.title('sobel_x', size=20)
plt.subplot(223)
edges_x = filters.sobel_v(im)
plt.imshow(edges_x)
plt.title('sobel_y', size=20)
plt.subplot(224)
edges = filters.sobel(im)
plt.imshow(edges)
plt.title('sobel', size=20)
plt.show()
```

```python
import numpy as np
from PIL import Image, ImageOps
import os 
import matplotlib.pylab as plt
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import filters
import cv2
```

```python
path = 'val/'
new_path = 'val_new/'
for name_file in os.listdir(path):
    for img_file in os.listdir(path+name_file):
        img = ImageOps.grayscale(Image.open(path+name_file+'/'+img_file))
        print(path+name_file+'/'+img_file)
        img = filters.sobel(img) * 255
            
        if not os.path.exists(new_path+name_file+'/'):
            os.makedirs(new_path+name_file+'/')
                
        cv2.imwrite(new_path+name_file+'/'+img_file, img)
```
